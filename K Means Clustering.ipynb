```python
X,y = make_blobs(n_samples=500,n_features=2,centers=5,random_state=3)
```

```python
plt.scatter(X[:,0],X[:,1],c =y)
```

```python
k = 5
clusters ={}
colors = ['yellow','blue','green','pink','orange']

for i in range(k):
    points = []
    center = 10*(2*np.random.random((X.shape[1],))-1)
    cluster ={
        'center':center,
        'points':points,
        'colors':colors[i]
    }
    clusters[i]= cluster
```

```python
print(clusters)
```

```python
def dist(x1,x2):
    return np.sqrt(np.sum((x1-x2)**2))
```

```python
def assign_points_cluster(clusters):
    
    for i in range(X.shape[0]):
        distance = []
        curr_x = X[i]
        for kx in range(k):
            d = dist(curr_x,clusters[kx]['center'])
            distance.append(d)


        curr_cluster = np.argmin(distance)
        clusters[curr_cluster]['points'].append(curr_x)
```

```python
def update_cluster(clusters):#M Steps
    for kx in range(k):
        pts = np.array(clusters[kx]['points'])
        if pts.shape[0]>0:
            new_center = pts.mean(axis=0)
            clusters[kx]['center']=new_center
            clusters[kx]['points']=[]
def plot_cluster(clusters):
    
    for kx in range(k):
        #print(len(clusters[kx][ 'points']))
        
        pts = np.array([clusters[kx]['points']])
        #print(pts.shape)
        
        #Plot the points
        try:
            plt.scatter(pts[:,0],pts[:,1])
        except:
            pass
        
        #Plot the cluster center
        uk = clusters[kx]['center']
        plt.scatter(uk[0],uk[1],color='black',marker="*")
```

```python
assign_points_cluster(clusters)
plot_cluster(clusters)
```

```python
assign_points_cluster(clusters)
plot_cluster(clusters)
update_cluster(clusters)
```
